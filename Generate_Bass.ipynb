```python
def generate_music(repeats):

    def generate_sample(lstm_model, len_sample, creativity, start_sequence=None):
        if start_sequence is None:
            current_sequence = torch.zeros(1, 1, 77).cuda()
        else:
            current_sequence = start_sequence
            
        output_sequence = [current_sequence.data.squeeze(1)]
        hd = None

        for i in range(len_sample):
            op, hd = lstm_model(current_sequence, [1], hd)
            probs = nn.functional.softmax(op.div(creativity), dim=1)
            current_sequence = torch.multinomial(probs.data, 1).squeeze().unsqueeze(0).unsqueeze(1).float()
            output_sequence.append(current_sequence.data.squeeze(1))

        generated_sequence = torch.cat(output_sequence, dim=0).cpu().numpy()  
        return generated_sequence
    
                        #(modelo, notas, "creatividad", secuencia inicial)
    verse = generate_sample(model, 30, 0.82, None)
    chorus = generate_sample(model, 80, 1.4, None)

    #La cancion final empieza con un verso
    song = verse

    #Añadir cuantos versos o coros se deseen
    song = np.vstack((song, verse))

    for i in range (repeats):
        
        song = np.vstack((song, verse))
        song = np.vstack((song, verse))
        song = np.vstack((song, verse))
        song = np.vstack((song, verse))
        song = np.vstack((song, verse))
        song = np.vstack((song, verse))
        song = np.vstack((song, chorus))
        song = np.vstack((song, chorus))
        song = np.vstack((song, chorus))
        song = np.vstack((song, chorus))

    song = np.vstack((song, verse))
    song = np.vstack((song, verse))
    song = np.vstack((song, verse))
    song = np.vstack((song, verse))
    song = np.vstack((song, verse))
    song = np.vstack((song, verse))
    song = np.vstack((song, verse))
    song = np.vstack((song, verse))

    return song

final_sequence = generate_music(2)   
```

```python
#RITMO DE LA CANCION: A MÁS ALTO EL VALOR, MÁS LENTO IRÁ; A MÁS BAJO, MÁS RÁPIDO 
deltaTime = 0.7
notes_array = []
time = 0
rng = (21,98)
sample = [i.nonzero()[0] + rng[0] for i in final_sequence]

for i in range(len(sample)):
    for note in sample[i]:
      if note < 256:
        if i==0 or note not in sample[i-1]:
                              #Note ON
                              #message, note, time, velocity, channel
          notes_array.append(('note_on', note, time, 90, 0))
          time = 0
    time += int(deltaTime*250)
    for note in sample[i]:
      if note < 256:
        if i==len(sample)-1 or note not in sample[i+1]:
                              #Note OFF
                              #message, note, time, velocity, channel
          notes_array.append(('note_off', note, time, 0, 0))
          time = 0
```

```python
def create_midi_file(notes_array, output_file, scale):

    og_name,extension = os.path.splitext(output_file)
    counter = 1
    new_name = f'{og_name}{extension}'

    while os.path.exists(new_name):
         new_name = f'{og_name}{counter}{extension}'
         counter +=1

    midi = MidiFile()
    track = MidiTrack()
    midi.tracks.append(track)

    for message_type, note, time, velocity, channel in notes_array:
            if note in scale and (note in range (40,77)):
                message = Message(message_type, channel=channel, note=note, velocity=velocity, time=time)
                track.append(message)

    midi.save(new_name)
                                                #Cambiar por la escala musical deseada
create_midi_file(notes_array, 'output_bass.mid', SOL_scale)
```
